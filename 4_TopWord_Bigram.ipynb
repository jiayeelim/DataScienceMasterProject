```python
import numpy as np
import pandas as pd

df_facebook = pd.read_csv('/content/drive/MyDrive/MDS Project LimJiaYee/Code/Dataset/Project2/facebookProject2_dashboard.csv')
df = df_facebook[['cleaned', 'restaurant', 'predicted_sentiment_model1','predicted_sentiment_model2','predicted_sentiment_model3']]
```

```python
df
```

```python
import ast
# Convert string representation of lists to actual lists
df['cleaned_1'] = df['cleaned'].apply(ast.literal_eval)

# Explode the lists into separate rows
df_exploded = df.explode('cleaned_1')

unwanted_words = ['nur', 'tak', 'dah', "'"]
df_exploded = df_exploded[~df_exploded['cleaned_1'].isin(unwanted_words)]
df_exploded
```

```python
# Count occurrences grouped by 'group_column' and 'words_column'
word_counts = df_exploded.groupby(['restaurant', 'predicted_sentiment_model1','cleaned_1']).size().reset_index(name='count')

#word_counts

#Get top 25 occurrences for each group
top25_word_counts = word_counts.groupby(['restaurant', 'predicted_sentiment_model1']).apply(lambda x: x.nlargest(25, 'count')).reset_index(drop=True)
top25_word_counts
```

```python
top25_word_counts.to_csv('/content/drive/MyDrive/MDS Project LimJiaYee/Code/Dataset/Project2/topword.csv')
```

## Bigram

```python
df['cleaned_text'] = df['cleaned'].apply(lambda x: ' '.join(word.strip("[],'") for word in x.split()))
df
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams

def get_ngrams(text, n=2):
    text = str(text)
    n_grams = ngrams(text.split(), n)
    returnVal = []

    try:
        for grams in n_grams:
            returnVal.append(' '.join(grams))
    except(RuntimeError):
        pass

    #return ' '.join(returnVal).strip()
    return returnVal

df["bigrams"] = df["cleaned_text"].apply(get_ngrams, n=2)
df
```

```python
# Explode the lists into separate rows
df_exploded_bigram = df.explode('bigrams')

# Count occurrences grouped by 'group_column' and 'words_column'
bigram_counts = df_exploded_bigram.groupby(['restaurant', 'predicted_sentiment_model1','bigrams']).size().reset_index(name='count')

bigram_counts

#Get top 25 occurrences for each group
top25_bigram_counts = bigram_counts.groupby(['restaurant', 'predicted_sentiment_model1']).apply(lambda x: x.nlargest(25, 'count')).reset_index(drop=True)
top25_bigram_counts
```

```python
top25_bigram_counts.to_csv('/content/drive/MyDrive/MDS Project LimJiaYee/Code/Dataset/Project2/bigram.csv')
```
